4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‘@’.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
import string                                             

password = input("Your password: ")

def is_password_valid(password: str) -> bool:
    if ' ' in password:
        return False

    upper = False
    lower = False
    digit = False
    special = False

    for char in password:    
        if char.isupper(): 
            upper = True
        elif char.islower():
            lower = True
        elif char.isdigit():
            digit = True
        elif char in string.punctuation:
            special = True
    return upper and lower and digit and special
    
if is_password_valid(password):
    print('True')
else:
    print('False')
````

10. Create a function (s) counting all lower case, upper case, digits, and special symbols from a given string and print them out.
   
```python

   example_password = "P@#yn26at^&i5ve"
```

````python
s = "P@#yn26at^&i5ve"

lower_case = 0 
upper_case = 0
digit = 0
special_symbols = 0

for char in s:
    if char.islower():
        lower_case += 1
    elif char.isupper():
        upper_case += 1
    elif char.isdigit():
        digit += 1
    else:
        special_symbols += 1

print("Lower case -", lower_case)
print("Upper_case -", upper_case)
print("Digit -", digit)
print("Special_symbols-", special_symbols)

````
